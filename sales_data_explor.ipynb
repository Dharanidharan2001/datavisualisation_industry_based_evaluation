```python
salesdf=sales_data[['Sales_in_thousands', '__year_resale_value',
       'Price_in_thousands', 'Engine_size', 'Horsepower', 'Width', 'Length', 'Fuel_capacity',
       'Fuel_efficiency', 'Power_perf_factor']]
```

```python
df=pd.DataFrame(salesdf)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(0, inplace=True)
df
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaler=StandardScaler()
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
scaled_data
```

```python
pca=PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
scaled_data
```

```python
x_pca
```

```python
colors = np.random.rand(10)
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=salesdf['Sales_in_thousands'],cmap='plasma')
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
```

```python
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_,columns=salesdf.keys())
```

```python
import seaborn as sns
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='plasma',)
```
